```python
# Split the data set into independent (X) and dependent (Y / diagnosis) data sets
X= df.iloc[:,2:31].values
Y= df.iloc[:,1].values
```

```python
# %25 test, %75 training
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size= 0.25, random_state= 0)
Y_train = np.array(Y_train, dtype=int)
Y_test = np.array(Y_test, dtype=int)
```

```python
#Scaling the data 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
def models(X_train, Y_train):

    #Logistic Regression
    from sklearn.linear_model import LogisticRegression
    logis = LogisticRegression(random_state= 0)
    logis.fit(X_train, Y_train)
    
    # Desicion Tree
    from sklearn.tree import DecisionTreeClassifier
    tree = DecisionTreeClassifier(criterion= 'entropy',random_state= 0)
    tree.fit(X_train,Y_train)

    # Random Forest Classifier
    from sklearn.ensemble import RandomForestClassifier
    forest = RandomForestClassifier(n_estimators= 10, criterion='entropy',random_state= 0)
    forest.fit(X_train,Y_train)

    print("[0]Logistic Regression Training Accuracy:", logis.score(X_train, Y_train))
    print("[1]Desicion Tree Training Accuracy:", tree.score(X_train, Y_train))
    print("[2]Random Forest Classifier Training Accuracy:", forest.score(X_train, Y_train))

    return logis, tree, forest
```

```python
model = models(X_train, Y_train)
```

```python
# testing
from sklearn.metrics import confusion_matrix
for i in range(len(model)):
    print (f"Model {i}")
    cm = confusion_matrix(Y_test, model[i].predict(X_test))
    
    TP = cm[0][0]
    FP = cm[0][1]
    TN = cm[1][1]
    FN = cm[1][0]
    
    print(cm)
    print("Testing Accuracy: ", (TP+TN)/(FP+FN+TP+TN))
    print()
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

for i in range(len(model)):
    print (f"Model {i}")
    print(classification_report(Y_test, model[i].predict(X_test) ))
    print(accuracy_score(Y_test, model[i].predict(X_test) ))
    print()
```
